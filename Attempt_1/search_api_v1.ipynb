```python
# Set the API key and Search Engine ID
API_KEY = 'AIzaSyAfFEE8C_6Nsd46RLjTbGP9j_bIN5dRvkE'
SEARCH_ENGINE_ID = 'a1dd1c2b37d1c4294'

# Define the search query
query = 'General contractors Near Madison, WI'
```

```python
# Define the API endpoint
url = f'https://www.googleapis.com/customsearch/v1?key={API_KEY}&cx={SEARCH_ENGINE_ID}&q={query}'

response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the JSON response
    data = response.json()
    
    # Extract the search results
    items = data.get('items', [])
    
    # Print the search results
    for item in items:
        title = item.get('title')
        link = item.get('link')
        print(f'Title: {title}')
        print(f'Link: {link}')
        print()
else:
    print(f'Error: {response.status_code}')
```

```python
# Define the number of search results per page
num_results_per_page = 10

# Define the total number of search results to retrieve
total_results_to_retrieve = 10

# Calculate the number of pages needed to retrieve all the search results
num_pages = total_results_to_retrieve // num_results_per_page
if total_results_to_retrieve % num_results_per_page != 0:
    num_pages += 1

results = {}
r_l = []
# Make a GET request for each page of search results
for page in range(num_pages):
    # Calculate the start index for the current page
    start_index = page * num_results_per_page + 1

    # Define the API endpoint for the current page
    url = f'https://www.googleapis.com/customsearch/v1?key={API_KEY}&cx={SEARCH_ENGINE_ID}&q={query}&start={start_index}&num={num_results_per_page}'

    # Make a GET request to the API endpoint
    response = requests.get(url)

    # Check if the request was successful
    if response.status_code == 200:
        # Parse the JSON response
        data = response.json()

        # Extract the search results
        items = data.get('items', [])

        # Print the search results
        for item in items:
            print(item)
            title = item.get('title')
            link = item.get('link')
            print(f'Title: {title}')
            print(f'Link: {link}')
            print()
            results[title] = {'company': title, 'website': link }
            r_l.append(item)
    else:
        print(f'Error: {response.status_code}')
    
```
